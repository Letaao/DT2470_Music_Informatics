```python
# add your code below
import pydub
import matplotlib.pyplot as plt
import numpy as np
from pydub import AudioSegment

# Setting up the plot parameters
params = {
    'legend.fontsize': 'x-large',
    'figure.figsize': (10, 5),
    'axes.labelsize': 'x-large',
    'axes.titlesize': 'x-large',
    'xtick.labelsize': 'x-large',
    'ytick.labelsize': 'x-large'
}
plt.rcParams.update(params)

audio = AudioSegment.from_wav("NHU05011021.wav")

# Creates an equivalent version of this with the specified number of channels
if audio.channels > 1:
    audio = audio.set_channels(1)

# Get raw audio data as an array of samples
samples = np.array(audio.get_array_of_samples())
samples = (samples / 2**16) * 2

# Get sample rate
sample_rate = audio.frame_rate

# Define the portion of the audio
start_time = 0.6  # in seconds
end_time = 0.7  # in seconds

# Get the segment samples
start_sample = int(start_time * sample_rate)
end_sample = int(end_time * sample_rate)
segment = samples[start_sample:end_sample]

# Compute the Fourier transform for both window types
windows = {
    "Boxcar": np.ones(len(segment)),  # Boxcar window
    "Hann": np.hanning(len(segment))   # Hann window
}

nyquist = sample_rate / 2  # Nyquist frequency

for window_name, window in windows.items():
    windowed_segment = segment * window
    fft_result = np.fft.fft(windowed_segment)
    fft_magnitude = np.abs(fft_result[:len(fft_result) // 2]) # only positive frequencies (first half)
    fft_magnitude_db = 20 * np.log10(fft_magnitude) # dB scale
    freqs = np.fft.fftfreq(len(fft_result), d=1/sample_rate)[:len(fft_result) // 2] / 1000  
    plt.plot(freqs, fft_magnitude_db, label=f"{window_name} Window")

plt.xlim(0, nyquist / 1000)  # in kHz
plt.xlabel("Frequency (kHz)")
plt.ylabel("Magnitude (dB)")
plt.title("Magnitude Spectrum of Audio Segment")
plt.legend()
plt.grid()
plt.show()
```

4. For the first 10 seconds of your audio file (or shorter if it is not so long), compute and plot its dB magnitude short-time Fourier transform using a Hann window of duration 25 ms with a window hopsize of 10 ms, and an FFT size of 8192 samples (hint: see signal.stft). Do the same using a Hann window of duration 100 ms with a window hopsize of 10 ms. Appropriately label your axes with "Frequency (kHz)" and "Time (s)". The frequency axis must be in kiloHertz, and limited to 0 to 5 kHz. The time axis must be in seconds. Choose a colormap that you feel describes your personality (see https://matplotlib.org/3.1.1/tutorials/colors/colormaps.html). 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import stft
from pydub import AudioSegment

# Setting up the plot parameters
params = {
    'legend.fontsize': 'x-large',
    'figure.figsize': (10, 10),
    'axes.labelsize': 'x-large',
    'axes.titlesize': 'x-large',
    'xtick.labelsize': 'x-large',
    'ytick.labelsize': 'x-large'
}
plt.rcParams.update(params)

# Load the audio file
audio = AudioSegment.from_wav("NHU05011021.wav")

# Ensure the audio is mono
if audio.channels > 1:
    audio = audio.set_channels(1)

# Get raw audio data as an array of samples
samples = np.array(audio.get_array_of_samples())
samples = (samples / 2**16) * 2

# Get sample rate
sample_rate = audio.frame_rate

end_time = 10  # in seconds
end_sample = int(end_time * sample_rate)
samples = samples[:end_sample]

fft_size = 8192  # FFT size
hop_size = int(0.01 * sample_rate)  # 10 ms hop size # the number of samples

# STFT with 25 ms Hann window
window_length_25ms = int(0.025 * sample_rate)  # 25 ms window length
window_25ms = np.hanning(window_length_25ms)
frequencies_25ms, times_25ms, Zxx_25ms = stft(samples, fs=sample_rate, window=window_25ms, nperseg=window_length_25ms, noverlap=window_length_25ms - hop_size, nfft=fft_size)
# Convert to dB magnitude
Zxx_magnitude_db_25ms = 20 * np.log10(np.abs(Zxx_25ms) + 1e-8)  # avoid log(0)

# STFT with 100 ms Hann window
window_length_100ms = int(0.1 * sample_rate)  # 100 ms window length
window_100ms = np.hanning(window_length_100ms)
frequencies_100ms, times_100ms, Zxx_100ms = stft(samples, fs=sample_rate, window=window_100ms, nperseg=window_length_100ms, noverlap=window_length_100ms - hop_size, nfft=fft_size)
# Convert to dB magnitude
Zxx_magnitude_db_100ms = 20 * np.log10(np.abs(Zxx_100ms) + 1e-8) 

# Plot for 25 ms window
plt.subplot(2, 1, 1)
plt.pcolormesh(times_25ms, frequencies_25ms / 1000, Zxx_magnitude_db_25ms, shading='gouraud', cmap='plasma')
plt.ylim(0, 5)  # Limit frequency axis to 0 to 5 kHz
plt.xlabel('Time (s)')
plt.ylabel('Frequency (kHz)')
plt.title('STFT Magnitude (dB) with 25 ms Hann Window')
plt.colorbar(label='Magnitude (dB)')

# Plot for 100 ms window
plt.subplot(2, 1, 2)
plt.pcolormesh(times_100ms, frequencies_100ms / 1000, Zxx_magnitude_db_100ms, shading='gouraud', cmap='plasma')
plt.ylim(0, 5)  # Limit frequency axis to 0 to 5 kHz
plt.xlabel('Time (s)')
plt.ylabel('Frequency (kHz)')
plt.title('STFT Magnitude (dB) with 100 ms Hann Window')
plt.colorbar(label='Magnitude (dB)')

plt.tight_layout()
plt.show()
```

5. Describe some of the advantages and nackdelar of using short or long time windows for time-frequency analysis.

> We can see that the longer time window blurs in the time direction and the shorter time window blurs in the frequency direction. The longer windows will thus give more resolution for frequency. The shorter windows will give more resolution for timing.

Short Time Windows: Better time resolution, good for analyzing transient events and rapidly changing signals. Poor frequency resolution, increased sensitivity to noise.

Long Time Windows: Better frequency resolution, useful for identifying closely spaced spectral features and analyzing signals with stable spectral content. Poor time resolution, less effective at detecting transient events, less adaptive to rapidly changing signals.

For the first 10 seconds of your audio file (or shorter if it is not so long), use the [librosa package](https://github.com/librosa) to compute its Mel spectrogram using Hann windows of duration 25 ms with a window hopsize of 10 ms. Use 128 Mel bands and an FFT size of 8192 samples. Display the dB magnitude with reference to the max power observed, and limit your y-axis between 0 and 5 kHz. Use the same colormap as you used above. See https://github.com/librosa/librosa/blob/main/examples/LibROSA%20demo.ipynb for help. 

```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
from pydub import AudioSegment

# Setting up the plot parameters
params = {
    'legend.fontsize': 'x-large',
    'figure.figsize': (10, 5),
    'axes.labelsize': 'x-large',
    'axes.titlesize': 'x-large',
    'xtick.labelsize': 'x-large',
    'ytick.labelsize': 'x-large'
}
plt.rcParams.update(params)

audio = AudioSegment.from_wav("NHU05011021.wav")

if audio.channels > 1:
    audio = audio.set_channels(1)

samples = np.array(audio.get_array_of_samples())
samples = (samples / 2**16) * 2
sample_rate = audio.frame_rate

end_time = 10  # in seconds
end_sample = int(end_time * sample_rate)
samples = samples[:end_sample]

fft_size = 8192  # FFT size
hop_size = int(0.01 * sample_rate)  # 10 ms hop size
window_length = int(0.025 * sample_rate)  # 25 ms window length
n_mels = 128  # Number of Mel bands

# Librosa expects float values normalized between -1 and 1
samples = samples.astype(np.float32)
samples /= np.max(np.abs(samples))

# Compute the Mel spectrogram using librosa
# https://librosa.org/doc/latest/generated/librosa.feature.melspectrogram.html
mel_spectrogram = librosa.feature.melspectrogram(
    y=samples,
    sr=sample_rate,
    n_fft=fft_size,
    hop_length=hop_size,
    win_length=window_length,
    window='hann',
    n_mels=n_mels
)

# Convert to dB scale
mel_spectrogram_db = librosa.power_to_db(mel_spectrogram, ref=np.max)

# Plot the Mel spectrogram
plt.figure(figsize=(10, 5))
librosa.display.specshow(
    mel_spectrogram_db,
    sr=sample_rate,
    hop_length=hop_size,
    x_axis='time',
    y_axis='mel',
    fmax=5000,  # Limit frequency to 5 kHz
    cmap='plasma'
)
plt.colorbar(format='%+2.0f dB')
plt.title('Mel Spectrogram (25 ms Hann Window)')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (kHz)')
plt.show()
```

1. Write a function that will take in the samples of an audio file, a frame size in samples, a frame hop size in samples, and compute and return the number of waveform zero crossings in each frame. A waveform x[n] undergoes a zero crossing when sign(x[n]) and sign(x[n+1]) are different. You will have to slice x[n] into chunks of a specified size, and for each of those chunks, count the number of sign changes.

```python
# add your code below
import numpy as np

def count_zero_crossings(audio_samples, frame_size, hop_size):
    """
    frame_size (int): Number of samples in each frame.
    hop_size (int): Number of samples to hop between frames.
    
    """
    audio_samples = np.array(audio_samples)
    
    zero_crossing_counts = []
    
    # Iterate over frames
    for start in range(0, len(audio_samples) - frame_size + 1, hop_size):
        frame = audio_samples[start:start + frame_size]
        zero_crossings = np.sum(np.diff(np.sign(frame)) != 0)
        zero_crossing_counts.append(zero_crossings)
    
    return zero_crossing_counts
```

3. Write a function that will take in the samples of an audio file, a frame size in samples, a hop size in samples, and a sampling rate, and compute and return the spectral centroid of each frame. The spectral centroid of a rectangular window of audio $x[n]$ of length $N$ (even) is defined as 
$$ R_{0.5}(x) = \frac{\sum_{k=0}^{N/2+1} \frac{F_s k}{N} |X[k]|}{\sum_{k=0}^{N/2+1} |X[k]|} $$
where $X[k]$ is the DFT of $x[n]$, and $F_s$ is the sampling rate.

```python
# add your code below
import numpy as np

def compute_spectral_centroid(audio_samples, frame_size, hop_size, sample_rate):
    
    audio_samples = np.array(audio_samples)
    
    spectral_centroids = []
    
     # Iterate over frames
    for start in range(0, len(audio_samples) - frame_size + 1, hop_size):
        frame = audio_samples[start:start + frame_size]
        
        X = np.fft.rfft(frame) 
        magnitudes = np.abs(X) 
        
        freqs = np.fft.rfftfreq(frame_size, 1/sample_rate)
        
        if np.sum(magnitudes) > 0:  
            spectral_centroid = np.sum(freqs * magnitudes) / np.sum(magnitudes)
        else:
            spectral_centroid = 0
        
        spectral_centroids.append(spectral_centroid)
    
    return spectral_centroids
```

```python
# add your code below
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import stft
from pydub import AudioSegment

# Setting up the plot parameters
params = {
    'legend.fontsize': 'x-large',
    'figure.figsize': (10, 10),
    'axes.labelsize': 'x-large',
    'axes.titlesize': 'x-large',
    'xtick.labelsize': 'x-large',
    'ytick.labelsize': 'x-large'
}
plt.rcParams.update(params)

# Load the audio file
audio = AudioSegment.from_wav("NHU05011021.wav")

# Ensure the audio is mono
if audio.channels > 1:
    audio = audio.set_channels(1)

# Get raw audio data as an array of samples
samples = np.array(audio.get_array_of_samples())
samples = (samples / 2**16) * 2

# Librosa expects float values normalized between -1 and 1
samples = samples.astype(np.float32)
samples /= np.max(np.abs(samples))

# Get sample rate
sample_rate = audio.frame_rate

window_size = int(sample_rate * 0.025)
hop_size = int(sample_rate * 0.01)
n_fft = 8192

# Compute MFCCs
mfccs = librosa.feature.mfcc(y=samples, sr=sample_rate, n_mfcc=10, n_fft=n_fft, hop_length=hop_size, win_length=window_size)

# Only keep the first 10 seconds (or shorter if audio is shorter)
num_frames = int(10 * sample_rate / hop_size)
mfccs = mfccs[:, :num_frames]

# Plot the MFCCs
plt.figure(figsize=(12, 6))

# Display the MFCCs
librosa.display.specshow(mfccs, x_axis='time', sr=sample_rate, hop_length=hop_size)
plt.colorbar(format='%+2.0f dB')
plt.title('MFCCs (First 10 Seconds)')
plt.xlabel('Time (s)')
plt.ylabel('MFCC Coefficients')
plt.tight_layout()
plt.show()
```
